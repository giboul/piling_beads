```python
import sympy as sym
```

```python
x = sym.Symbol("x")
y = sym.Symbol("y")
z = sym.Symbol("z")
r = sym.Symbol("r")

x1 = sym.Symbol("x_1")
y1 = sym.Symbol("y_1")
z1 = sym.Symbol("z_1")
r1 = sym.Symbol("r_1")

x2 = sym.Symbol("x_2")
y2 = sym.Symbol("y_2")
z2 = sym.Symbol("z_2")
r2 = sym.Symbol("r_2")

x3 = sym.Symbol("x_3")
y3 = sym.Symbol("y_3")
z3 = sym.Symbol("z_3")
r3 = sym.Symbol("r_3")

s = sym.Symbol("s")

d1 = (x-x1)**2 + (y-y1)**2 + (z-z1)**2
d2 = (x-x2)**2 + (y-y2)**2 + (z-z2)**2
d3 = (x-x3)**2 + (y-y3)**2 + (z-z3)**2

display(sym.Equality(d1, (r+r1)**2))
display(sym.Equality(sym.collect(sym.expand(d1-d2), (x, y, z)), (r+r1)**2 - (r+r2)**2))
display(sym.Equality(sym.collect(sym.expand(d1-d3), (x, y, z)), (r+r2)**2 - (r+r3)**2))
```

```python
# Find linear solution (x(z), y(z)) with `z` as an independent variable
"""
2(x2-x1)     2(y2-y1)    2(z2-z1)    |   (r+r1)² - (r+r2)² - (x1²+y1²+z1²) + (x2²+y2²+z2²)
2(x3-x1)     2(y3-y1)    2(z3-z1)    |   (r+r1)² - (r+r3)² - (x1²+y1²+z1²) + (x3²+y3²+z3²)

    a           b      |   c
    d           e      |   f    

"""
a = +2*(x2-x1)
b = +2*(y2-y1)
c = -2*(z2-z1)*z + (r+r1)**2 - (r+r2)**2 - (x1**2 + y1**2 + z1**2) + (x2**2 + y2**2 + z2**2)

d = +2*(x3-x1)
e = +2*(y3-y1)
f = -2*(z3-z1)*z + (r+r1)**2 - (r+r3)**2 - (x1**2 + y1**2 + z1**2) + (x3**2 + y3**2 + z3**2)

y_z = (f*a - d*c)/(e*a - d*b)
x_z = c/a - b/a * y_z
```

```python
"""
(x-x1)² + (y-y1)² + (z-z1)² = (r+r1)²
as
(α+βz)² + (γ+δz)² + (z-z1)² = (r+r1)²
"""
beta, alpha = sym.poly(x_z - x1, z).coeffs()
delta, gamma = sym.poly(y_z - y1, z).coeffs()
```

```python
"""
az² + bz + c = 0
a = β² + δ² + 1
b = 2(αβ + γδ - z1)
c = α² + γ² + z1² - (r+r1)²
"""
a = beta**2 + delta**2 + 1
b = 2*(alpha*beta + gamma*delta - z1)
c = alpha**2 + gamma**2 + z1**2 - (r+r1)**2

det = b**2 - 4*a*c

z_sol =  (-b + s*(det)**0.5) / (2*a)
```

```python
display(x_z.subs(dict(z=z_sol.subs(dict(s=-1)))))
display(y_z.subs(dict(z=z_sol.subs(dict(s=+1)))))
```

```python
display(x_z.subs(dict(z=z_sol)).together())
```
